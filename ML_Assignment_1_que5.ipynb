```python
import pandas as pd

df = pd.read_csv('/content/WineQT.csv', sep=',')

df['quality'] = df['quality'].apply(lambda x: 0 if x <= 6 else 1)

# print(df['quality'])

cols_to_normalize = df.columns.difference(['quality'])

# Normalize using min-max scaling
for col in cols_to_normalize:
    min_val = df[col].min()
    max_val = df[col].max()
    df[col] = (df[col] - min_val) / (max_val - min_val)


print(df.tail())
```
